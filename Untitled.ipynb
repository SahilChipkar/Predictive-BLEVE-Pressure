```python
# Import necessary libraries
import pandas as pd

# Load the training and testing datasets
train_data = pd.read_csv('train.csv')



from sklearn.preprocessing import OneHotEncoder
# Convert categorical feature "Status" using one-hot encoding
encoder = OneHotEncoder(sparse=False)
status_encoded = encoder.fit_transform(train_data[["Status"]])  # Encode the "Status" feature in the training data
status_encoded_df = pd.DataFrame(status_encoded, columns=encoder.get_feature_names(["Status"]))
train_data_encoded = pd.concat([train_data.drop(columns=["Status"]), status_encoded_df], axis=1)


X = train_data_encoded .drop(columns=['Target Pressure (bar)'])
y = train_data_encoded ['Target Pressure (bar)']

```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_percentage_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor

svr = SVR()
svr.fit(X_train, y_train)
```

```python
# 3. Random Forest Regression
rf_reg = RandomForestRegressor()
rf_reg.fit(X_train, y_train)
```

```python
# 4. XGBoost Regression
xgb_reg = XGBRegressor()
xgb_reg.fit(X_train, y_train)
```

```python
y_test_lin_reg = lin_reg.predict(X_test)
mape_lin_reg = mean_absolute_percentage_error(y_test, y_test_lin_reg)
r2_lin_reg = r2_score(y_test, y_test_lin_reg)
```

```python
y_test_svr = svr.predict(X_test)
mape_svr = mean_absolute_percentage_error(y_test, y_test_svr)
r2_svr = r2_score(y_test, y_test_svr)
```

```python
y_test_rfg = rf_reg.predict(X_test)
mape_rfg = mean_absolute_percentage_error(y_test, y_test_rfg)
r2_rfg = r2_score(y_test, y_test_rfg)
```

```python
y_test_xgb = xgb_reg.predict(X_test)
mape_xgb = mean_absolute_percentage_error(y_test, y_test_xgb)
r2_xgb = r2_score(y_test, y_test_xgb)
```

```python
ML_Model = ['Liner regression','Suport vector','Random Forest' ,'XgBoost' ]
MAPE = [mape_lin_reg,mape_svr,mape_rfg,mape_xgb]
R2_Score =[r2_lin_reg,r2_svr,r2_rfg,r2_xgb]
#creating dataframe
result = pd.DataFrame({ 'ML Model' : ML_Model,
                         'MAPE' :   MAPE,
                      'R2 Score' : R2_Score })
```

```python
sorted_result=result.sort_values(by=['R2 Score'],ascending=False).reset_index(drop=True)
```
